```python
import pandas as pd

def convert(cell):
    if cell == "n.a.":
        return 'shreyas nayak'
    return cell


def convert2(cell):
    if cell == -1:
        return ("0");
    return cell

df=pd.read_excel("stock_data.xlsx","Sheet1",na_values=["not available","n.a."],converters = {
    'people': convert,
    'eps': convert2
})
df
```

```python
# df.to_excel("new.xlsx",sheet_name="stocks")
df.to_excel("new.xlsx",sheet_name="stocks",startrow= 1,startcol=2,index=False)
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx')as writer:
    df_stocks.to_excel(writer,sheet_name="stock")
    df_weather.to_excel(writer,sheet_name="weather")
```
